3. Створити дві змінні, що посилаються на будь-які ціле число та число з рухомою комою та продемонструвати такі арифметичні операції: додавання, віднімання, ділення, множення, ділення без залишку, ділення по модулю, приведення до ступеню. Всі результати операцій вивести на екран.

```python

a = 10    # (int)
b = 3.5   # (float)

# Арифметичні операції
print("Додавання:", a + b)         # 10 + 3.5 = 13.5
print("Віднімання:", a - b)        # 10 - 3.5 = 6.5
print("Множення:", a * b)          # 10 * 3.5 = 35.0
print("Ділення:", a / b)           # 10 / 3.5 = 2.857142857142857
print("Ділення без залишку:", a // b)  # 10 // 3.5 = 2.0 (заокруглюється до меншого цілого)
print("Ділення по модулю:", a % b) # 10 % 3.5 = 3.0 (остача від ділення)
print("Піднесення до степеня:", a ** b) # 10 ** 3.5 ≈ 3162.28
```

15. Порівняйте список та кортеж. Назвіть схожості та відмінності, випадки використання.

Схожості:

Індексовані: і списки, і кортежі підтримують доступ до елементів за допомогою індексів.

Ітерабельні: можна використовувати їх у циклах (наприклад, for), оскільки вони ітеровані.

Можуть містити елементи різних типів: 
як у списках, так і в кортежах можна зберігати елементи різних типів (цілі числа, рядки, списки, кортежі тощо).

Відмінності:

Змінність:

Список — змінний тип даних, тобто ви можете змінювати його елементи (додавати, видаляти, змінювати).
Кортеж — незмінний тип даних, тобто після створення його елементи не можна змінювати. Це гарантує, що дані в кортеже не будуть змінюватися під час виконання програми.
Синтаксис:

Список створюється за допомогою квадратних дужок: list_example = [1, 2, 3].
Кортеж створюється за допомогою круглих дужок: tuple_example = (1, 2, 3) (для кортежу з одним елементом необхідно використовувати кому: (1,)).

Продуктивність:

Кортежі можуть бути більш швидкими в операціях, пов'язаних з доступом до елементів, порівняно зі списками, через їхню незмінність.
Списки можуть мати вищу витрату пам'яті, оскільки вони повинні зберігати додаткову інформацію для підтримки змінюваності.

Методи:

Списки мають багато вбудованих методів для маніпуляції даними, таких як append(), remove(), pop(), sort(), extend(), і так далі.

Кортежі мають тільки методи count() та index(), оскільки вони не можуть бути змінені після створення.

Випадки використання:

Список зазвичай використовується, коли потрібно працювати з колекцією елементів, яку можна змінювати: додавати нові елементи, змінювати наявні, видаляти елементи.

Приклад використання списку:

Зберігання і обробка динамічних даних (наприклад, список покупок, список студентів в групі).
Кортеж зазвичай використовується, коли потрібно зберігати колекцію елементів, яка не повинна змінюватися. Це дозволяє гарантувати, що дані залишаються постійними, що може бути корисним для захисту від випадкових змін або помилок.

Приклад використання кортежу:

Ключі в словниках (оскільки кортежі є незмінними і можуть бути використані як ключі в словниках).
Коли потрібно зберігати невеликі, незмінні групи даних (наприклад, координати точки: (x, y)).

Висновок:
Списки підходять для динамічних даних, де необхідно змінювати елементи.
Кортежі використовуються для збереження постійних даних або коли потрібно гарантувати, що дані не будуть змінюватися.

# VI. Множини (Set).

17. Створити множину без елементів. Після цого за допомогою методу додайте кілька різних елементів до множини. Чи множини є змінним типом даних?

```python
# Створюємо порожню множину
empty_set = set()

# Додаємо елементи до множини
empty_set.add(1)
empty_set.add(3.14)
empty_set.add("Hello")
empty_set.add((1, 2))

# Виведення результату
print("Множина після додавання елементів:", empty_set)

# Перевірка, чи є множина змінним типом
is_mutable = isinstance(empty_set, set)
print("Чи є множина змінним типом даних?", is_mutable)
```

Множини - змінний тип даних, адже ми можемо додавати елементи після створення множини, і вона змінюється в процесі роботи.

18. Створити множину, маючи список my_list = [1, 1, 2, 67, 67, 8, 9]. Пояснити, чому "зникли" деякі елементи.

```python
# Список
my_list = [1, 1, 2, 67, 67, 8, 9]

# Перетворюємо список на множину
my_set = set(my_list)

# Виведення результату
print("Множина:", my_set)
```

За замовчуванням всі елементи множини - унікальні, тож створивши список з повторюваними елементами і перетворивши список на множину, отримаємо список з унікальною інформацією, а саме -- множину. 

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
# Створюємо дві множини
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Операція обʼєднання (union)
union_result = set1.union(set2)

# Операція різниці (difference)
difference_result = set1.difference(set2)

# Операція пересічення (intersection)
intersection_result = set1.intersection(set2)

# Операція симетричної різниці (symmetric difference)
symmetric_difference_result = set1.symmetric_difference(set2)

# Виведення результатів
print("Обʼєднання:", union_result)
print("Різниця (set1 - set2):", difference_result)
print("Пересічення:", intersection_result)
print("Симетрична різниця:", symmetric_difference_result)
```

20. Створіть пустий словник. До нього додайте чотири пари елементів такі, щоб їхні ключі були різних типів. Чи може список бути ключем? Чому?

```python
my_dict = {}

my_dict["name"] = "Alice"  # Ключ рядок
my_dict[1] = "One"         # Ключ ціле число
my_dict[3.14] = "Pi"       # Ключ число з рухомою комою
my_dict((1, 2)) = "Tuple"  # Ключ кортеж

# Виведення результату
print("Словник:", my_dict)
```

Ні, список не може бути ключем в словнику, тому що список є змінним типом даних, а ключі в словнику повинні бути незмінними типами даних. Ключі повинні бути хешованими, а список, як змінний об'єкт, не підтримує хешування

21. Створіть словник, де значенням в одній з пар теж буде словник, який теж має вкладений словник. Виведіть на екран значення, що міститься в словнику, що знаходиться на найнижчому рівні ієрархії вкладеності (найбільш внутрішній).
